```python
import numpy as np
import pandas as pd
import seaborn as sn
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import random
```

```python
bounds = []
midpoints = []
for i in range(5):
  a = 0
  b = 0
  a = np.random.randint(0, 90)
  b = np.random.randint(a+10, 100)
  bounds.append([a, b])
  midpoints.append((a+b)/2)

print(bounds)
print(midpoints)
  
```

```python
important_indices = []

important_indices.append(np.random.randint(0, 4))
important_indices.append(np.random.randint(important_indices[0]+1, 5))


```

```python
data = dict()

n = 1000  # Size of the Data
features = ['x1', 'x2', 'x3', 'x4', 'x5']  # List of classes
# x1 = list(np.random.randint(low=-100, high=100, size=n))
# x2 = list(np.random.randint(low=-100, high=100, size=n))
for i in range(5):
    data[features[i]] = list(np.random.randint(bounds[i][0], bounds[i][1], n))

data['y'] = list()

for i in range(n):
    weights = None
    if data[features[important_indices[0]]][i] > midpoints[important_indices[0]] and data[features[important_indices[1]]][i] > midpoints[important_indices[1]]:
        weights = [0.2, 0.8]
    elif data[features[important_indices[0]]][i] < midpoints[important_indices[0]] and data[features[important_indices[1]]][i] < midpoints[important_indices[1]]:
        weights = [0.8, 0.2]
    else:
        weights = [0.5, 0.5]

    data['y'].append((random.choices([0, 1], weights=weights))[0])

# x3 = list(np.random.randint(low = -100,high=100,size=n))
data = pd.DataFrame(data)
data
# sn = list(range(500))
```
